# SECTION 1: Coverage Alignment Check

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("../data/processed")

# Load required datasets
camps_df = pd.read_csv(DATA_PATH / "mobile_camp_locations.csv")
gap_df   = pd.read_csv(DATA_PATH / "district_gap_analysis.csv")

print("camps_df shape:", camps_df.shape)
print("gap_df shape:", gap_df.shape)
```

```python
predicted_capacity = camps_df['coverage_population'].sum()
estimated_unreached = gap_df['unreached_population'].sum()

alignment_ratio = predicted_capacity / estimated_unreached * 100

print("\nðŸ“Š COVERAGE ALIGNMENT CHECK (Capacity-Based)")
print(f" Total estimated unreached demand : {estimated_unreached:,.0f}")
print(f" Total proposed service capacity : {predicted_capacity:,.0f}")

print(f"\n Alignment ratio: {alignment_ratio:.2f}%")

if alignment_ratio >= 80 and alignment_ratio <= 120:
    print(" âœ… Capacity is well-aligned with estimated demand")
elif alignment_ratio < 80:
    print(" âš ï¸ Capacity may be insufficient â€” consider more camps")
else:
    print(" âš ï¸ Capacity exceeds estimated demand â€” possible overlap or over-allocation")
```

# SECTION 2: Geographic Fairness Check

```python
state_camps = camps_df.groupby('state').size()
state_unreached = gap_df.groupby('state')['unreached_population'].sum()

correlation = state_camps.corr(state_unreached)
print(f"State allocation correlation: {correlation:.3f}")
```

# SECTION 3: Cost Benchmark Validation

```python
cost_per_person = camps_df['total_cost'].sum() / camps_df['coverage_population'].sum()
BENCHMARK = 100  # â‚¹100 per enrolment

print(f"Cost per enrolment: â‚¹{cost_per_person:.2f}")
print(f"Benchmark: â‚¹{BENCHMARK}")
print("Status:", "âœ… Within benchmark" if cost_per_person <= BENCHMARK else "âŒ Above benchmark")
```
